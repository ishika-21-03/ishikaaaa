```python
import pandas as pd
```

```python
df = pd.read_csv("ipl_matches_smalldata.csv")
```

```python
top_10_batter = df.sort_values(by='total_runs', ascending=False).head(10)
```

```python
mean_total_runs = top_10_batter['total_runs'].mean()
mode_total_runs = top_10_batter['total_runs'].mode()
```

```python
print(f"Mean total_runs scored by top 10 batter: {mean_total_runs}")
print(f"Mode total_runs scored by top 10 batter: {mode_total_runs.values}")
```

```python
batter_with_500_plus_total_runs = df[df['total_runs'] > 500]
```

```python
print(batter_with_500_plus_total_runs)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,7))
plt.bar(top_10_batter["batter"], top_10_batter["total_runs"])
plt.xlabel("Batter")
plt.ylabel("Total_runs")
plt.title("Top 10 batter by total_runs")
plt.show()
```
